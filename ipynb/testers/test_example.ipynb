```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
qc = QuantumCircuit(3, 3) #* Quantum circuit w/ 3 qubits, 3 classical bits
qc.x(1) #* Pauli-X on q1
qc.h(range(3)) #* Hademard on all
qc.cx(0, 1) #* CNOT on 0 --> 1
qc.measure_all() #* convert to classical format
qc.draw(output="mpl")
```
